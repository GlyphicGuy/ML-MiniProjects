```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = sns.load_dataset('healthexp')
```

```python
df.columns
```

```python
def convdv(string):
  var = 0

  if string == 'USA':
    var=1
  elif string == 'Canada':
    var=2
  elif string == 'Japan':
    var=3
  elif string == 'Germany':
    var=4
  elif string == 'France':
    var=5
  elif string == 'Great Britain':
    var=6

  return var
```

```python
convdv('USA')
```

```python
# converting the country to numeric values
df['Country'] = df['Country'].apply(convdv)
```

```python
df
```

```python
x = df.drop('Life_Expectancy',axis=1)
print(x)
```

```python
y = df['Life_Expectancy']
print(y)
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
LR = LinearRegression()
#creating the model
```

```python
LR.fit(xtrain,ytrain)
#this is training the model
```

```python
sv = np.array([2060,5,5000]).reshape(1,-1)
```

```python
LR.predict(sv)
```
