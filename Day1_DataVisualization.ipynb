# Hotel Bill Calculator

```python
def slvHotel(bf, lunch, dineer):
  bill = 0
  if(bf == 'SouthIndian'):
    bill+= 50
  elif(bf == 'NorthIndian'):
    bill+= 70

  if(lunch == 'Chinese'):
    bill+= 200
  elif(lunch == 'Korean'):
    bill+= 250

  if(dineer == 'Veg'):
    bill+= 100
  elif(dineer == 'NonVeg'):
    bill+= 150

  return bill
```

```python
iface2 = gr.Interface(
    fn=slvHotel,
    inputs=[gr.Radio(['SouthIndian','NorthIndian']),gr.Radio(['Chinese','Korean']),gr.Radio(['Veg','NonVeg'])],
    outputs=gr.Text(label="Your Bill")
)
```

```python
iface2.launch()
```
